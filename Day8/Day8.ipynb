### Advent of Code Day 8: Seven Segment Search

```python
uniquePatterns = []
outputPatterns = []   # Parsed values after "|"

with open("./input", 'r') as input:
    for line in input:
        uniquePattern, output = line.split("|")
        outputPattern = output.split(" ")[1:]
        outputPattern[-1] = outputPattern[-1].rstrip("\n")

        outputPatterns.append(outputPattern)
        uniquePatterns.append(uniquePattern.split(" ")[:-1])
```

```python
for pattern, output in zip(uniquePatterns, outputPatterns):
    print(pattern, output)
```

**Part One** : Count digit values with unique number of segments over output digits (count how many times do 1, 4, 7, or 8 appear over output values)

```python
count = 0
for line in outputPatterns:
    for value in line:
        if len(value) in [2,3,4,7]: # Corresponding seven segment digit: 1, 7, 4 or 8
            count += 1

count
```

```python
patternDicts = []

for patterns in uniquePatterns:
    patternDicts.append(dict())          # {lenght: pattern list}

    for pattern in patterns:
        if len(pattern) not in patternDicts[-1]: patternDicts[-1][len(pattern)] = [set(pattern)]
        else: patternDicts[-1][len(pattern)].append(set(pattern))

patternDicts
```

```python
letterValues = {letter : None for letter in 'abcdefg'}
trainDict = dict()

for pattern in "acedgfb cdfbe gcdfa fbcad dab cefabd cdfgeb eafb cagedb ab".split(" "):
    if len(pattern) not in trainDict: trainDict[len(pattern)] = [set(pattern)]
    else: trainDict[len(pattern)].append(set(pattern))

print(trainDict)
```

Let $s(x)$ denote the set of 7 segments conforming the digit $x$, we can classify each digit by considering its set cardinal:


- Two segments digit: $ s(1) = \{c, f\}$

- Three segments digit: $ s(7) = \{a, c, f\}$

- Four segments digit: $ s(4) = \{b, c, d, f \}$

- Five segments digit: $ s(2) = \{a, c, d, e, g\}$, $ s(3) = \{a, c, d, f, g\}$, $ s(5) = \{a, b, d, f, g\}$

- Six segments digit: $ s(6) = \{a, b, d, e, f, g\}$, $s(9) = \{a, b, c, d, f, g\}$

- Seven segments digit: $ s(8) = \{a, b, c, d, e, f, g\}$

$\{a\} = s(7) - s(1) $

```python
def getA(patternDict, letterValues): 
    set1 = patternDict[2][0]
    set4 = patternDict[3][0]

    for letter in 'abcdefg': 
        if letter in set4 - set1: 
            letterValues['a'] = letter
```

```python
getA(trainDict, letterValues)
letterValues
```

$\{d\} = s(4) \cap s(2) \cap s(3) \cap s(5)$

```python
def getD(patternDict, letterValues):
    set4 = patternDict[4][0]
    
    for letter in set4:
        intersection = True

        for set in patternDict[5]:
            intersection &= letter in set

        if intersection: letterValues['d'] = letter
```

```python
getD(trainDict, letterValues)
letterValues
```

$\{b\} = s(4) - \{d \} - s(1)$

```python
def getB(patternDict, letterValues):
    set4 = patternDict[4][0]
    set1 = patternDict[2][0]

    for letter in set4:
        if letter != letterValues['d'] and letter not in set1: 
            letterValues['b'] = letter
```

```python
getB(trainDict, letterValues)
letterValues
```

$\{c\} = s(9) - s(5)$

```python
def getCAndF(patternDict, letterValues):
    set1 = patternDict[2][0]

    for set in patternDict[5]: 
        if letterValues['b'] not in set:
            if len(set1.intersection(set)) == 1:
                set2 = set

    for cValue in set1.intersection(set2): 
        letterValues['c'] = cValue

    for fValue in set1:
        if fValue != cValue: 
            letterValues['f'] = fValue
```

```python
getCAndF(trainDict, letterValues)
letterValues
```

```python
def getEAndG(patternDict, letterValues):
    for set in patternDict[6]:
        # Avoid 6 and 0 digits
        if letterValues['c'] in set and letterValues['d'] in set:
            set9 = set


    set8 = patternDict[7][0]

    for letter in set8 - set9:
        letterValues['e'] = letter
    
    for lastLetter in 'abcdefg':
        if lastLetter not in letterValues.values():
            letterValues['g'] = lastLetter
```

```python
getEAndG(trainDict, letterValues)
letterValues
```

```python
def getLetterValues(patternDict, letterValues):
    """ Frankenstein decoder """
    getA(patternDict, letterValues)
    getD(patternDict, letterValues)
    getB(patternDict, letterValues)
    getCAndF(patternDict, letterValues)
    getEAndG(patternDict, letterValues)
```

```python
digitValues = [set("abcefg"), set("cf"), set("acdeg"), set("acdfg"), set("bcdf"), set("abdfg"), set("abdefg"), set("acf"), set("abcdefg"), set("abcdfg")]
getLetterValues(trainDict, letterValues)
print(letterValues)

def getValue(pattern, letterValues):
    inverseValues = {v: k for k, v in letterValues.items()}
    original_pattern = "".join([inverseValues[letter] for letter in pattern])
    return original_pattern

def decoder(pattern, letterValues):
    originalPattern = getValue(pattern, letterValues)
    return digitValues.index(set(originalPattern))
```
